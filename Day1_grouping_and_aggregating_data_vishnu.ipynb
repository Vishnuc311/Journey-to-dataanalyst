### Example:

**Dataset**:

I'm using a sample dataset with information about different regions, products, and the corresponding sales.

```python
import pandas as pd

# Sample dataset
data = {'Region': ['East', 'West', 'East', 'West', 'East'],
        'Product': ['A', 'B', 'A', 'B', 'C'],
        'Sales': [100, 200, 150, 300, 120]}

df = pd.DataFrame(data)
df
```

***Grouping by a Single Column***

Task: Group by Region and calculate total sales.

```python
# Grouping by Region and summing Sales
grouped_data = df.groupby('Region')['Sales'].sum()
grouped_data
```

Analysis:

East Region: Total sales are 370.

West Region: Total sales are 500.


***Grouping by a Single Column with Multiple Aggregations***

Task: Apply both sum() and mean() to Sales for each region.

```python
# Applying sum and mean together
agg_data = df.groupby('Region')['Sales'].agg(['sum', 'mean'])
agg_data
```

Analysis:

East Region: Total sales are 370 with an average of 123.33.

West Region: Total sales are 500 with an average of 250.00.


***Grouping by Multiple Columns***

Task: Group by Region and Product, then calculate both sum() and mean() for Sales.

```python
# Grouping by Region and Product, and applying multiple aggregations
grouped_by_multiple = df.groupby(['Region', 'Product'])['Sales'].agg(['sum', 'mean'])
grouped_by_multiple
```

Analysis:

East Region, Product A: Total sales are 250 with an average of 125.

East Region, Product C: Total sales are 120 with an average of 120.

West Region, Product B: Total sales are 500 with an average of 250.
